```python
'''Problem 

1. year has many non-year values and it is in boject so need to convert it to int
2. price has ask price and need to conver dtype to int
3. kms_driven object to int and also has nan values 
4. fuel type also has nan values 
5. name so will keep first 3 words of name to improve data'''
```

```python
backup = df.copy()
```

```python
df['year']
```

```python
df['year'].str.isnumeric()
```

```python
df = df[df['year'].str.isnumeric()]
```

```python
df['year'] = df['year'].astype(int)
```

```python
df = df[df['Price'] != "Ask For Price"]
```

```python
# Remove commas from the 'Price' column and then convert to integer
df['Price'] = df['Price'].str.replace(',', '').astype(int)
```

```python
df['kms_driven'] = df['kms_driven'].str.split(' ').str.get(0)
```

```python
df['kms_driven'] = df['kms_driven'].str.replace(',','')
```

```python
df = df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven'] = df['kms_driven'].astype(int)
```

```python
df = df[~df['fuel_type'].isna()]
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['name'] = df['name'].str.split()
```

```python
df['name'] = df['name'].str.slice(0,3)
```

```python
df.head()
```

```python
df['name'] = df['name'].str.join(' ')
```

```python
df.head()
```

```python
df = df.reset_index(drop = True)
```

```python
df.describe()
```

```python
df = df[df['Price']<6e6].reset_index(drop = True)
```

```python
x = df.drop(columns = 'Price')
y = df['Price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
oe = OneHotEncoder()
oe.fit(x[['name', 'company', 'fuel_type']])
```

```python
oe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories = oe.categories_ ),['name','company','fuel_type']),remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
r2_score(y_test, y_pred)
```

```python
scores = []
for i in range(10):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state = i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans, lr)
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    scores.append(r2_score(y_test, y_pred))
```

```python
import numpy as np
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state = np.argmax(scores) )
lr = LinearRegression()
pipe = make_pipeline(column_trans, lr)
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)
r2_score(y_test, y_pred)
```
